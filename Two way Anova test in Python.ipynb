## Question


* Null hypothesis H0: Salary offered at placement is not affected by educational background and  work exeperiance
* Alternate hypothesis H1: Salary offered at placement is affected by educational background and  work exeperiance

```python
df = pd.read_csv(r"C:\Users\Samir\Downloads\Placement_Data_Full_Class.csv")
```

```python
df
```

```python
from statsmodels.formula.api import ols
```

```python
#Variable name = ols('dependent_variable~indeoendent_variables')
ano2 = ols('salary ~ degree_t+workex',data=df).fit()
```

```python
Two_way_anova =sm.stats.anova_lm(ano2,type="III")
```

```python
Two_way_anova
```

##### Analyse the result:

* For both independent variables the p-value turn out to be greater than 0.05 hence we would accepting the null hypothesis. This implies that we have sufficient proof to say that salary offered in placement is not affected by educational background and work exeperiance.
